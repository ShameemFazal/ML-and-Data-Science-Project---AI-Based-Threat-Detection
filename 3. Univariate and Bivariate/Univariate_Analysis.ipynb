```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Recommended setup
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)

# Loading   dataset 
df = pd.read_csv("User_and_Admin_Activity_Data_1.csv")
```

# Used value counts or bar plots:

```python
 categorical_cols = ['RecordType', 'Operation', 'UserId', 'ResultStatus', 'Workload','GeoLocation', 'ApplicationDisplayName', 'LogonError'
]


for col in categorical_cols:
    plt.figure()
    sns.countplot(data=df, x=col, order=df[col].value_counts().index)
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```

# Univariate Analysis  Time Features & Custom Flags

```python
# Numeric-Like Columns (e.g., HourOfDay)

# HourOfDay
sns.histplot(df['HourOfDay'], bins=24, kde=True)
plt.title("Distribution of HourOfDay")
plt.show()

# DayOfWeek
sns.countplot(x='DayOfWeek', data=df)
plt.title("Distribution of DayOfWeek")
plt.show()

# Weekend Flag
sns.countplot(x='IsWeekend', data=df)
plt.title("Weekend vs Weekday")
plt.show()
```

# Calculating the covariance between numerical columns 

```python
# Calculate covariance
cov_value1 = df['RecordType'].cov(df['HourOfDay'])
print(f"Covariance between RecordType and HourOfDay: {cov_value1}")

cov_value2 = df['IsLoginFailure'].cov(df['IsRiskyHour'])
print(f"Covariance between IsLoginFailure and IsRiskyHour: {cov_value2}")
```

# Calculating the correlation between numerical columns 

```python
# Calculate correlation
corr_value1 = df['RecordType'].corr(df['HourOfDay'])
print(f"Correlation between RecordType and HourOfDay: {corr_value1}")

corr_value2 = df['IsLoginFailure'].corr(df['IsRiskyHour'])
print(f"Correlation between IsLoginFailure and IsRiskyHour: {corr_value2}")
```
