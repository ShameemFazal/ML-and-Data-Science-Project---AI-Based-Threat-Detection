```python
'''
Read Data from CSV
Split independent / dependent
Split Train set /Test set
Preprocessing – standard Scalar
Grid Search Cross Validation 
Model Creation - RANDOM FOREST CLASSIFIER
Confusion matrix
Classification report
ROC_AUC_SCORE value
'''
```

```python
import pandas as pd
```

```python
dataset = pd.read_csv("Feature_Importance_Columns.csv")
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['Operation_FileAccessed', 'Operation_FileModified',
       'Operation_FileUploaded',
       'Operation_FileDownloaded', 'Operation_MoveToDeletedItems',
       'IsRiskyHour', 'Operation_UserLoginFailed', 'GeoLocation_IND',
       'ClientIP_203.28.245.168', 'ClientIP_192.168.1.100','ResultStatus_unknown']]
```

```python
dependent=dataset[['IsThreat']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
X_train
```

```python
Y_train
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

param_grid = {'criterion':['gini','entropy'],
              'max_features': ['auto','sqrt','log2'],
              'n_estimators':[10,100]}
grid = GridSearchCV(RandomForestClassifier(), param_grid, refit = True, verbose=3,n_jobs=-1,scoring='f1_weighted')

# fitting the model for grid search
grid.fit(X_train, Y_train)
```

```python
re=grid.cv_results_

grid_predictions = grid.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,grid_predictions)
```

```python
from sklearn.metrics import classification_report
clf_report = classification_report(Y_test,grid_predictions)
```

```python
print(clf_report)
```

```python
print(cm)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(Y_test,grid.predict_proba(X_test)[:,1])
```
